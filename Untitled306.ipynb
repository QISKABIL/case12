```python
import pandas as pd
import matplotlib.pyplot as plt
sampath=pd.read_csv('energy.csv')
print(sampath)

#ANDHRA KARNATAKA
#LINE CHART
srh=sampath[sampath['State']=='Andhra']
rcb=sampath[sampath['State']=='Karnataka']

fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='yellow')
diag[1].plot(rcb['Date'],rcb['Consumption'])

srh=sampath[sampath['State']=='Telengana']
rcb=sampath[sampath['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='yellow')
diag[1].bar(rcb['Date'],rcb['Consumption'])

#PIE CHART
state1='Kerala'
state2='Delhi'
tdp=[state1,state2]

srh=sampath[sampath['State']=='Kerala']['Consumption'].sum()
rcb=sampath[sampath['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]

fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp,autopct='%1.1f%%')





```
