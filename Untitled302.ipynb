```python
import pandas as venu
import matplotlib.pyplot as ijaz

lr=venu.read_csv('energy.csv')
print(lr)

#LINE CHART
# Andhra , Karnataka
srh=lr[lr['State']=='Andhra']
rcb=lr[lr['State']=='Karnataka']

fig,diag=ijaz.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])

#Telengana, Tripura
srh=lr[lr['State']=='Telengana']
rcb=lr[lr['State']=='Tripura']

fig,diag=ijaz.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='green')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='red')

# pie chart
csds='Kerala'
cse='Delhi'
ece=[csds,cse]

tdp=lr[lr['State']=='Kerala']['Consumption'].sum()
ysr=lr[lr['State']=='Delhi']['Consumption'].sum()
jsp=[tdp,ysr]

fig,diag=ijaz.subplots(2)
diag[0].pie(jsp,labels=ece,autopct='%1.1f%%')











```
